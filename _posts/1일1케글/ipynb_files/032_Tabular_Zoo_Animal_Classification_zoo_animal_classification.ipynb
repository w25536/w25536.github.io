```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("uciml/zoo-animal-classification")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import os

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler

from sklearn.model_selection import train_test_split

import tensorflow as tf
```

```python
df = pd.read_csv(os.path.join(path, "zoo.csv"))
```

```python
df.info()
```

```python
df = df.drop(["animal_name"], axis=1)
```

```python
{column: list(df[column].unique()) for column in df.columns}
```

```python
y = df["class_type"]
X = df.drop(["class_type"], axis=1)
```

```python
scaler = MinMaxScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
```

```python
from sklearn.linear_model import LogisticRegression


log_model = LogisticRegression()
log_model.fit(X_train, y_train)


from sklearn.svm import SVC

svm_model = SVC(C=1.0)
svm_model.fit(X_train, y_train)

from sklearn.neural_network import MLPClassifier

nn_model = MLPClassifier(hidden_layer_sizes=(16, 16))
nn_model.fit(X_train, y_train)
```

```python
log_acc = log_model.score(X_test, y_test)
svm_model = svm_model.score(X_test, y_test)
nn_model = nn_model.score(X_test, y_test)

print("Accuracy Reslts")
print("Logistic Model:", log_acc)
print("SVM Model", svm_model)
print("Neural Network Model", nn_model)
```
